# Country Clusters

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

Load data from the csv file: <i> 'Countries.csv'</i>.


```python
raw_data = pd.read_csv('Countries.csv')
raw_data.head(10)
```

## Plot the data

Plot the <i>'Longtitude'</i> and <i>'Latitude'</i> columns. 

```python
# Scatter plot using 'Longitude' and 'Latitude' resembling the world map
plt.scatter(raw_data['Longitude'], raw_data['Latitude'])
plt.xlim(-180, 180)
plt.ylim(-90,90)
plt.show()
```

## Select the features

Create a copy of that data and remove all parameters apart from <i>Longitude</i> and <i>Latitude</i>.

```python
data = raw_data.copy()
data = data.drop(['name'], axis=1)
data.head()
```

## Clustering

Assume there are only two clusters. 

```python
kmeans = KMeans(2)
```

```python
# this code will apply k-means clustering with 2 clusters to x
kmeans.fit(data)
```

### Clustering Resutls

```python
identified_clusters = kmeans.fit_predict(data)
data_with_clusters = raw_data.copy()
data_with_clusters['Cluster'] = identified_clusters
data_with_clusters.head(10)
```

```python
plt.scatter(data_with_clusters['Longitude'], data_with_clusters['Latitude'],
           c=data_with_clusters['Cluster'],cmap='rainbow')
plt.xlim(-180, 180)
plt.ylim(-90,90)
plt.show()
```
